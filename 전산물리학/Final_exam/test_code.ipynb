```python
import numpy as np
from sklearn.utils import shuffle
import matplotlib.pyplot as plt

yourid = 20224531
path = r"C:\github\Data\final_files\p1data.npy"

with open(path,'rb') as infile:
    test_images = np.load(infile)
    test_labels = np.load(infile)

# Shuffle the data
test_images, test_labels = shuffle(test_images, test_labels, random_state=yourid)

# Identify unique labels
unique_labels = np.unique(test_labels)

# Find the first index for each label
first_indices = {label: np.where(test_labels == label)[0][0] for label in unique_labels}

# Plot the images corresponding to the first index of each label
plt.figure(figsize=(10, 10))
for i, label in enumerate(unique_labels):
    idx = first_indices[label]
    plt.subplot(1, len(unique_labels), i + 1)
    plt.imshow(test_images[idx], cmap='gray')
    plt.title(f'Label {label}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
