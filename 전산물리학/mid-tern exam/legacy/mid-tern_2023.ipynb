A lowrider is doing its signature motion in a festival. Using their rear wheels as pivotal points, it lifts its front part of the car to the high point and pumps it as the front wheels hit the ground so that the car stands up almost vertically.  Please check the [Youtube movies.](https://www.youtube.com/watch?v=VZnjfFFedKA) (4:00~5:00)

a) The car is initially at rest.  
b) The car starts to jump by the repulsive force on the tires.  
c). The car continues to jump as it bounces on the ground.  
d) It reaches to the highest point at some point.  
e). Now, it goes down again and after several repulsive jumps, it stops eventually partly due to friction and air drag and partly due to decreasing repulsion.

```python
import numpy as np
import matplotlib.pyplot as plt

# Car parameters
m = 1500  # Mass of the car (kg)
k = 50000  # Spring constant of the suspension (N/m)
c = 5000  # Damping coefficient (N/(m/s))
g = 9.81  # Acceleration due to gravity (m/s^2)

# Initial conditions
x0 = 0.0  # Initial position (m)
v0 = 0.0  # Initial velocity (m/s)

# Time parameters
t0 = 4.0  # Start time (s)
t_end = 5.0  # End time (s)
dt = 0.001  # Time step (s)

# Function to calculate the net force acting on the car
def net_force(x, v):
    if x <= 0:
        return -k * x - c * v - m * g
    else:
        return -m * g

# Numerical integration using Euler's method
t = np.arange(t0, t_end, dt)
x = np.zeros_like(t)
v = np.zeros_like(t)

x[0] = x0
v[0] = v0

for i in range(1, len(t)):
    a = net_force(x[i-1], v[i-1]) / m
    v[i] = v[i-1] + a * dt
    x[i] = x[i-1] + v[i] * dt

# Plot the results
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Lowrider Motion')

plt.subplot(2, 1, 2)
plt.plot(t, v)
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')

plt.tight_layout()
plt.show()
```
