```python
# Bubble Sort

import tkinter as tk
import random
import time

def bubble_sort(data):
    for i in range(len(data)-1):
        for j in range(len(data)-i-1):
            if data[j] > data[j+1]:
                data[j], data[j+1] = data[j+1], data[j]
                draw_bars(data)
                time.sleep(0.01)
                root.update()

def draw_bars(data):
    canvas.delete("all")
    canvas_width = canvas.winfo_width()
    canvas_height = canvas.winfo_height()
    bar_width = canvas_width / len(data)
    scale = canvas_height / max(data)
    for i, value in enumerate(data):
        x1 = i * bar_width
        y1 = canvas_height
        x2 = x1 + bar_width
        y2 = canvas_height - (value * scale)
        canvas.create_rectangle(x1, y1, x2, y2, fill="blue")
    canvas.update()

def on_closing():
    root.destroy()

root = tk.Tk()
root.title("Bubble Sort Visualization")
root.protocol("WM_DELETE_WINDOW", on_closing)  # 창 닫기 이벤트 처리

canvas = tk.Canvas(root, width=600, height=400)
canvas.pack()

data = [random.randint(10, 100) for _ in range(50)]

bubble_sort(data)

root.mainloop()
```

```python
import tkinter as tk
import random
import time

def insertion_sort(data):
    for i in range(1, len(data)):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
            draw_bars(data)
            time.sleep(0.01)
            root.update()
        data[j + 1] = key
        draw_bars(data)
        time.sleep(0.01)
        root.update()

def draw_bars(data):
    canvas.delete("all")
    canvas_width = canvas.winfo_width()
    canvas_height = canvas.winfo_height()
    bar_width = canvas_width / len(data)
    scale = canvas_height / max(data)
    for i, value in enumerate(data):
        x1 = i * bar_width
        y1 = canvas_height
        x2 = x1 + bar_width
        y2 = canvas_height - (value * scale)
        canvas.create_rectangle(x1, y1, x2, y2, fill="blue")
    canvas.update()

def on_closing():
    root.destroy()

root = tk.Tk()
root.title("Insertion Sort Visualization")
root.protocol("WM_DELETE_WINDOW", on_closing)

canvas = tk.Canvas(root, width=600, height=400)
canvas.pack()

data = [random.randint(10, 100) for _ in range(50)]

insertion_sort(data)

root.mainloop()
```
